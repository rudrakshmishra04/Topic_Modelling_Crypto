```python
import pandas as pd
```

```python
# Import dataset
df = pd.read_csv(r"C:\Users\domdt\Desktop\MPS Data Analytics\Predictive Analytics\Team Project\Final04202023_V2.csv")
```

```python
# Selecting Columns to carry forward
df = df[['id',
 'score',
 'created',
 'sentimentTextBlob',
 'polarityTextBlob',
 'sentimentAFINN',
 'polarityAFINN',
 'sentimentVader',
 'polarityVader',
 'weightedScore',
 'weightedpolarityTextBlob',
 'weightedpolarityAFINN',
 'weightedpolarityVader']]
```

```python
df['weightedpolarityTextBlob'] = df['weightedpolarityTextBlob'] *100
df['weightedpolarityAFINN'] = df['weightedpolarityAFINN'] *100
df['weightedpolarityVader'] = df['weightedpolarityVader'] *100
```

```python
# Grouping columns on created and taking sum
grouped_df = df.groupby("created")
```

```python
mean_df = grouped_df.sum()
```

```python
mean_df
```

```python
# Import Bitcoin Dataset
df2 = pd.read_csv(r"C:\Users\domdt\Desktop\MPS Data Analytics\Predictive Analytics\Team Project\BitcoinData_Oct_Jan_edited.csv")
```

```python
# Reset Index
mean_df.reset_index(drop=False, inplace=True)
```

```python
mean_df
```

```python
# Join the two data sets on the column "created"
df3 = pd.merge(mean_df, df2, how='inner', on='created',
         left_index=False, right_index=False, sort=True,)
```
